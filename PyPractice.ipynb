```python
n = int (input ("Enter the value of n:"))
def fibonacci(n):
  if n <= 0:
    return 0
  elif n == 1:
    return 1
  else:
    return fibonacci(n-1) + fibonacci(n-2)

# print the first n numbers of the Fibonacci series
for i in range(n):
  print(fibonacci(i))
```

```python
# Prime number function 

def primes(numbers):
    prime = []
    for num in numbers:
        if num > 1:
            for i in range(2, num):
                if (num % i) == 0:
                    break
            else:
                prime.append(num)
    return prime
```

```python
numbers = [2, 3, 4, 5, 6, 7, 8, 9, 10]
prime= primes(numbers)
print(prime)
```
