```python
#Seaborn

import pandas as pd
import matplotlib.pyplot as plt 
import seaborn as sns 
```

```python
df= sns.load_dataset('tips')
```

```python
df
```

```python
plt.figure(figsize=(8,6))
df["size"].value_counts().plot(kind="pie",autopct="%1.1f%%")
```

```python
sns.stripplot(y = 'tip',data =df, jitter= True)
plt.ylabel('tip($)')
plt.show()
```

```python
df.columns
```

```python
#Grouping with stripplot()
plt.figure(figsize=(8,7))
sns.stripplot(x='size',y='tip',data=df)
plt.ylabel('tip($)')
plt.show()
```

```python
#Grouping with Boxplot()
plt.figure(figsize=(8,7))
sns.boxplot(x='size',y='tip',data=df)
plt.ylabel('tip($)')
plt.show()
```

```python
sns.stripplot(x='time',y='tip',data=df,size=5,jitter=True)
plt.ylabel('tip($)')
plt.show()
```

```python
sns.boxplot(x='time',y='tip',data=df)
plt.ylabel('tip($)')
plt.show()
```

```python
sns.stripplot(x='day',y='tip',hue="sex",data=df,size=5,jitter=True)
plt.ylabel('Tip($)')
plt.show()
```

```python
sns.boxplot(x='day',y='tip',hue="sex",data=df)
plt.ylabel('Tip($)')
plt.show()
```

```python
sns.boxplot(x='smoker',y='tip',data=df)
plt.ylabel('Tip($)')
plt.show()
```

```python
sns.boxplot(x='sex',y='tip',data=df)
plt.ylabel('Tip($)')
plt.show()
```

```python
sns.swarmplot(y='tip',data=df)
plt.ylabel('tip($)')
plt.show()
```

```python
plt.figure(fidsize=(20,10),facecolor="lightgreen")
x=np.linespace(-4,+4,50)

plt.subplot(4,1,1)
plt.plot(x,x**2)
plt.xlabel("X axis",size =15)
plt.ylabel("Y axis",size=15)
plt.title("Plot 1",size=20)

plt.subplot(4,1,2)
plt.plot(x,np.tan(x))
plt.xlabel("X axis",size=15)
plt.ylabel("Y axis",size=15)
plt.title("Plot 2",size=20)

plt.subplot(4,1,3)
plt.plot(x,np.sin(x))
plt.xlabel("X axis",size= 15)
plt.ylabel("Y axis",size=15)
plt.title("Plot 3", size=20)

plt.subplot(4,1,3)
plt.plot(x,np.cos(x))
plt.xlabel("X axis",size= 15)
plt.ylabel("Y axis",size=15)
plt.title("Plot 4", size=20)
```
